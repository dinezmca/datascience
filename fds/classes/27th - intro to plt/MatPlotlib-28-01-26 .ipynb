```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x1 = np.linspace(0,10,50)
x1
```

```python
%matplotlib inline
x1 = np.linspace(0, 10, 50)

# create a plot figure
#fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
#plt.plot(x1, np.tan(x1), '--')
plt.show()
```

```python
x1 = np.linspace(0, 10, 5) # Only 5 points
y = np.cos(x1)
print(f"x1 values - {x1}")
print(f'y values - {y}')

plt.plot(x1, y, 'o-') # 'o-' shows the dots and the lines
plt.title("Sine Wave with only 5 points")
plt.show()
```
