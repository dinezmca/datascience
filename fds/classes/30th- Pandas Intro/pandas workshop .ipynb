```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([[1, 2], 
                 [3, 4]])

# Axis 0: Down the columns (1+3 and 2+4)
print(np.sum(data, axis=0))  # Output: [4, 6]

# Axis 1: Across the rows (1+2 and 3+4)
print(np.sum(data, axis=1))  # Output: [3, 7]

plt.plot(data)
plt.show()
```
