```python

import pandas as pd
import numpy as np
df = pd.read_excel('Rawdata.xlsx')
df.head()

```

```python
df['Name'] = df['Name'].str.replace("\W", "", regex=True)

df['Name']

```

```python
df['Age'] =df['Age'].str.extract('(\\d+)')


df['Age'] = df['Age'].fillna(np.mean(pd.to_numeric(df['Age'])))
df['Age']
```
